```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
LEGEND_SIZE=15
LABEL_SIZE=18
TICK_SIZE=15
```

```python
x = [20, 40, 60, 80]
ostk_m_gemma = [0.14, 0.26, 0.37, 0.39]
ostk_m_llama = [0.14, 0.32, 0.53, 0.57]
ostk_m_gpt4o = [0.23, 0.46, 0.76, 0.74]
ostk_l_gemma = [0.18, 0.29, 0.33, 0.37]
ostk_l_llama = [0.1, 0.22, 0.46, 0.45]
ostk_l_gpt4o = [0.26, 0.52, 0.61, 0.66]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(16, 4))
ax1.plot(x, ostk_m_gpt4o, label='gpt4o', color="black", marker='^')
ax1.plot(x, ostk_m_gemma, label='gemma', color="blue", marker='o')
ax1.plot(x, ostk_m_llama, label='llama 3.1', color="red", marker='s')
ax1.set_title("Metric Queries", fontsize=LABEL_SIZE)
ax1.set_ylabel("Accuracy", fontsize=LABEL_SIZE)
fig.supxlabel("Percentage of samples used", fontsize=LABEL_SIZE, y=-0.05)
ax2.plot(x, ostk_l_gpt4o, color="black", marker='^')
ax2.plot(x, ostk_l_gemma, color="blue", marker='o')
ax2.plot(x, ostk_l_llama, color="red", marker='s')
ax2.set_title("Log Queries",fontsize=LABEL_SIZE)
ax2.set_ylabel("F-1 Score", fontsize=LABEL_SIZE)
ax1.tick_params(labelsize=TICK_SIZE)
ax2.tick_params(labelsize=TICK_SIZE)
ax1.set_ylim([0,0.81])
fig.legend(fontsize=LEGEND_SIZE, loc='center', bbox_to_anchor=(0.62, 0.75), ncol=1)
plt.savefig("ostk_ft_accuracy.pdf", dpi=1200, bbox_inches="tight")
plt.show()
```
